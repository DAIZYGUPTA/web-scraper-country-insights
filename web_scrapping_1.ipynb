# Built a Python-based web scraper to extract structured country-level data including population, capital, and area from a public website.

```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
url = "https://www.scrapethissite.com/pages/simple/"

headers = {"User-Agent": "Mozilla/5.0"}

res = requests.get(url, headers = headers)
if res.status_code == 200:
    with open ("web_scrapping_1_original.html" , 'w' , encoding="utf-8") as f:
        f.write(res.text)
else :
    print("Error in scrapping")
```

```python
with open ("web_scrapping_1_original.html" , "r" , encoding="utf-8") as file:
    html_content = file.read()
```

```python
soup = BeautifulSoup(html_content, "lxml")
```

```python
countries = soup.select("div.country")
```

```python
countries
```

```python
data = []
for each in countries:
    name_tag = each.select_one("h3.country-name")
    capital_tag = each.select_one("span.country-capital")
    population_tag = each.select_one("span.country-population")
    area_tag = each.select_one("span.country-area")

    name = name_tag.text.strip() if name_tag else None
    capital = capital_tag.text.strip() if capital_tag else None
    population = int(population_tag.text.strip()) if population_tag else None 
    area = float(area_tag.text.strip()) if area_tag else None

    data.append({
        "Name" : name,
        "Capital" : capital,
        "Population" : population,
        "Area" : area
    })
```

```python
data
```

```python
df = pd.DataFrame(data)
```
